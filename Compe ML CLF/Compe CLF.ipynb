```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data\\train.csv',index_col='Id')
```

```python
df.head()
```

```python
df['age']=df['age'].str.replace('y','').astype(int)
mean_valueMI=df['MonthlyIncome'].mean()
mean_valueND=df['NumberOfDependents'].mean()
df['MonthlyIncome'].fillna(value=mean_valueMI, inplace=True)
df['NumberOfDependents'].fillna(value=round(mean_valueND,0), inplace=True)
```

```python
df_1=df[df['SeriousDlqin2yrs']==1]
```

```python
df_1.info()
```

```python
df_2=df[df['SeriousDlqin2yrs']==0]
```

```python
df_2.info()
```

```python
df_2S=df_2.sample(n=4856)
```

```python
df3=pd.concat([df_2S,df_1])
```

```python
df3
```

```python
df.head()
```

```python
X = df3.drop(columns=['SeriousDlqin2yrs'])
y = df3['SeriousDlqin2yrs']
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.value_counts(normalize=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
from sklearn import decomposition
sc = StandardScaler()
pca = decomposition.PCA()
X=df.iloc[:,:-1]
n_components = list(range(1,X.shape[1]+1,1))
dtreeCLF = DecisionTreeClassifier()
```

```python
reg_log = Pipeline(steps = [
    ("imputer", SimpleImputer()),
    ("scaler", StandardScaler()),
    ("reglog", LogisticRegression())
])
reg_log_param = {
    "imputer__strategy": ['mean', 'median'],
    "reglog__penalty": ['l1', 'l2'],
    "reglog__C": np.logspace(0, 4, 10)
}

rand_forest = RandomForestClassifier()
rand_forest_param = {
    "n_estimators": [10, 100, 1000],
    "max_features": [1,2,3],
    'criterion' : ['entropy']
}


svm = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("selectkbest", SelectKBest()),
    ("svm", SVC())
])


svm_param = {
    'selectkbest__k': [2, 3, 4],
    'svm__kernel': ['linear', 'rbf', 'sigmoid', 'poly'],
    'svm__C': [0.001, 0.1, 0.5, 1, 5, 10, 100],
    'svm__degree': [1,2,3,4],
    'svm__gamma': ['scale', 'auto']
}

knn = Pipeline(steps = [    
    ("knn_scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])

knn_param = {
    'knn__leaf_size': [30,60,90],
    'knn__n_neighbors':[5,25,50,100],
    'knn__p' : [2,4,6]

    
}

Dtree = Pipeline(steps = [ 

    ("sc", sc),
    ("pca", pca),
    ("DtreeCLF", DecisionTreeClassifier())
])

Dtree_param = {
    "DtreeCLF__criterion": ["gini", "entropy"], 
    "DtreeCLF__max_depth": [1,2,3,4,5,6,7, None],
    "DtreeCLF__min_samples_leaf": [1, 2, 3]
}



```

```python
gs_reg_log = GridSearchCV(reg_log,
                         reg_log_param,
                         cv = 10,
                         scoring = 'roc_auc',
                         verbose = 1,
                         n_jobs = -1)

gs_rand_forest = GridSearchCV(rand_forest,
                         rand_forest_param,
                         cv = 10,
                         scoring = 'roc_auc',
                         verbose = 1,
                         n_jobs = -1)

gs_svm = GridSearchCV(svm,
                         svm_param,
                         cv = 10,
                         scoring = 'roc_auc',
                         verbose = 1,
                         n_jobs = -1)
gs_knn = GridSearchCV(knn,
                         knn_param,
                         cv = 10,
                         scoring = 'roc_auc',
                         verbose = 1,
                         n_jobs = -1)
gs_Dtree = GridSearchCV(Dtree,
                         Dtree_param,
                         cv = 10,
                         scoring = 'roc_auc',
                         verbose = 1,
                         n_jobs = -1)





grids = {#"gs_reg_log": gs_reg_log,
        "gs_rand_forest": gs_rand_forest,
        #"gs_svm": gs_svm,
        #"gs_knn": gs_knn,
        #"gs_dtree": gs_Dtree   
        
        }
```

```python
for nombre, grid_search in grids.items():
    grid_search.fit(X_train, y_train)
```

```python
print(gs_rand_forest.best_score_)
print(gs_rand_forest.best_params_)
print(gs_rand_forest.best_estimator_)
```

```python
preds = gs_rand_forest.best_estimator_.predict(X_test)
roc_auc_score(y_test, preds)
```
